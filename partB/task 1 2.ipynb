# Task 1.2 - Key Assumptions

**Assumption 1**  
Assumption: Time series subsequences are comparable after Z-normalization; Why the method needs it: both DTW distances and lower bounds are defined on normalized values to capture shape.  
Violation scenario: Domains where absolute level is the signal (for example, threshold alarms) so normalization removes key information.  
Paper reference: Section 4.2.1 (online mean/variance and Z-normalization).

**Assumption 2**  
Assumption: A relatively small Sakoe-Chiba warping window R is appropriate; Why the method needs it: envelopes and constrained DTW depend on R to define admissible alignments.  
Violation scenario: Time series with large timing distortions where the optimal path lies outside the window.  
Paper reference: Section 4.1 and 4.2 (DTW with window and envelope construction).

**Assumption 3**  
Assumption: Lower bounds (LB_Kim and LB_Keogh) are tight enough to prune most candidates; Why the method needs it: the claimed speedups depend on pruning before full DTW.  
Violation scenario: Highly repetitive or near-constant series where bounds are loose and pruning is minimal.  
Paper reference: Section 4.1.2 to 4.1.4 and Section 4.2 (UCR Suite optimization cascade).
