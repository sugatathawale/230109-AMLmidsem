# Task 2.1 - Dataset Selection and Setup

I use a synthetic time-series dataset because the paper targets subsequence search under DTW, which is naturally defined on continuous sequences rather than tabular features.  
The dataset is a random-walk series T (length 2000) with a sine-wave query Q (length 128) embedded once with small noise.  
This is a reasonable testbed because it preserves the core task: exact DTW subsequence search with a known ground-truth location.  
The series yields 1,873 candidate subsequences (2000 - 128 + 1), which satisfies the 100+ sample requirement for this time-series task.  
Compared to the paper's massive random-walk datasets, this toy dataset is tiny and has only one embedded motif, so it is unsuitable for stress-testing scalability but ideal for validating correctness and the pruning cascade.

**Preprocessing:** each candidate subsequence and the query are Z-normalized before comparison, following the paper's setup.

```

import numpy as np

SEED = 42
rng = np.random.default_rng(SEED)

m = 128
T_len = 2000
x = np.linspace(0, 2 * np.pi, m)
Q = np.sin(2 * x) + 0.1 * rng.normal(size=m)
base = rng.normal(0, 0.6, T_len).cumsum()

embed_start = 700
T = base.copy()
T[embed_start:embed_start + m] = Q + 0.05 * rng.normal(size=m)

np.savez("partB/data/synthetic_series.npz", T=T, Q=Q, embed_start=embed_start)

print(f"Saved synthetic_series.npz with T_len={len(T)}, Q_len={len(Q)}, embed_start={embed_start}")
```
