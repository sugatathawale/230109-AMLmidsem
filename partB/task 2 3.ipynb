# Task 2.3 - Result, Comparison, and Reproducibility Checklist


**Achieved result (toy dataset):** UCR-DTW found the best match at index 700 with distance 0.5053 and mean search time 0.0791s (3 runs).  
**Baseline:** Naive DTW took 0.3364s on the same dataset.

**Paper comparison (Table 2, |Q|=128 random walk):** UCR-DTW = 0.159s for 1 million points, SOTA-DTW = 2.447s.  
**Why the gap exists:** My dataset is far smaller (2,000 points vs 1,000,000), uses a single embedded motif, and is run in Python without low-level optimizations. The paper's results use highly optimized C code and large-scale random walk data. The relative trend still matches: pruning yields a clear speedup even on a toy dataset.


```

import matplotlib.pyplot as plt

labels = ['Naive DTW', 'UCR-DTW']
values = [0.3364, 0.0791]

plt.figure(figsize=(5, 3))
plt.bar(labels, values)
plt.ylabel('Seconds (mean of 3 runs)')
plt.tight_layout()
plt.savefig('partB/results/reproduction_runtime.png', dpi=160)
print('Saved: partB/results/reproduction_runtime.png')
```
