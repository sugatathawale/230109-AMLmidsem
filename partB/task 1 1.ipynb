# Task 1.1 - Core Contribution / Architecture

**Paper:** Searching and Mining Trillions of Time Series Subsequences under Dynamic Time Warping (KDD 2012)


**Step-by-Step Method Description**

**Step 1: Normalize the query (Q)**  
- **Description:** Z-normalize the query to remove offset and scale, so comparisons are shape-based.  
- **Reference:** Section 4.2.1 (Early Abandoning Z-Normalization), equations for mean and variance.  
- **Purpose:** Ensures DTW comparisons are invariant to mean/variance and supports tight lower bounds.

**Step 2: Build a warping envelope for Q**  
- **Description:** For each time index, compute upper and lower bounds within a Sakoe-Chiba window R.  
- **Reference:** Section 4.1.2 (Lower Bounding), Figure 4 (LB_Keogh).  
- **Purpose:** Enables the LB_Keogh lower bound to prune candidates cheaply.

**Step 3: Slide a window over the long series T to generate candidates C**  
- **Description:** For each subsequence of length |Q|, compute running mean/variance for fast Z-normalization.  
- **Reference:** Section 4.2.1 (online mean/variance formulas).  
- **Purpose:** Avoids re-normalizing from scratch for every candidate.

**Step 4: Apply LB_KimFL (O(1))**  
- **Description:** Use first/last points (simplified LB_Kim) to quickly reject obvious mismatches.  
- **Reference:** Section 4.1.2 (LB_KimFL).  
- **Purpose:** Extremely cheap filter before more expensive bounds.

**Step 5: Apply LB_Keogh (Q-envelope)**  
- **Description:** Compute distance from C to the Q-envelope; early abandon if exceeds best-so-far.  
- **Reference:** Section 4.1.2 and 4.1.3 (LB_Keogh plus early abandoning).  
- **Purpose:** Prunes most candidates without full DTW.

**Step 6: Apply a second LB_Keogh (C-envelope)**  
- **Description:** Build envelope on C and bound Q against it (tighter cascade).  
- **Reference:** Section 4.2 (UCR Suite optimizations) and Table 1 algorithm outline.  
- **Purpose:** Further tightens the lower bound before DTW.

**Step 7: Compute constrained DTW with early abandoning**  
- **Description:** If lower bounds pass, compute DTW with window R, abandoning if partial cost exceeds best-so-far.  
- **Reference:** Section 4.1.4 (Early Abandoning of DTW), Figure 6.  
- **Purpose:** Exact distance while keeping computation minimal.

**Step 8: Update best-so-far (BSF) and continue**  
- **Description:** Keep the lowest DTW distance and index of the best match.  
- **Reference:** Table 1 (UCR-DTW outline).  
- **Purpose:** Enables pruning using BSF for subsequent candidates.

**Final Summary Sentence:** The paper solves exact subsequence search under DTW at massive scale, claiming that a cascade of tight lower bounds plus early abandoning and online normalization makes exact DTW search faster than prior state-of-the-art and even faster than Euclidean search on large datasets.
