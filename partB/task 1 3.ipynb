# Task 1.3 - What the Paper Claims to Improve


**Baseline / prior method:** The paper's main comparison is against SOTA-DTW (state-of-the-art DTW subsequence search that uses Z-normalization, LB_Keogh, and early abandoning).  

**Baseline limitation identified:** Even with LB_Keogh and early abandoning, SOTA-DTW remains too slow for very large datasets (millions to trillions of points).  

**How the paper overcomes it:** The UCR Suite adds a cascade of additional optimizations (LB_KimFL, a second LB_Keogh, early abandoning of normalization, and DTW early abandoning) that collectively prune more candidates and reduce per-candidate cost.  

**When the method might not outperform:** On short or highly repetitive series where lower bounds are loose, the extra bound computations add overhead but do not prune much, so runtime can approach or even exceed SOTA-DTW.
